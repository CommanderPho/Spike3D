```python
from pyphoplacecellanalysis.External.pyqtgraph import QtGui, QtCore

class DoubleVerticalLines(pg.GraphicsObject):
    def __init__(self, x_positions, gap=2):
        super().__init__()
        self.x_positions = x_positions
        self.gap = gap
        self.generatePicture()

    def generatePicture(self):
        self.picture = QtGui.QPicture()
        p = QtGui.QPainter(self.picture)
        p.setPen(pg.mkPen('w'))
        for x in self.x_positions:
            p.drawLine(QtCore.QPointF(x, 0), QtCore.QPointF(x, 1))
            p.drawLine(QtCore.QPointF(x + self.gap, 0), QtCore.QPointF(x + self.gap, 1))
        p.end()

    def paint(self, p, *args):
        p.drawPicture(0, 0, self.picture)

    def boundingRect(self):
        return QtCore.QRectF(self.picture.boundingRect())
```

```python
from pyphocorehelpers.indexing_helpers import BinningContainer

## Instantaneous peak firing rates for each cell can be used as a measure of the population ordering during an epoch. Might be a better measure than first/median spike during laps and longer epochs:

replay_instantaneous_time_bin_size_seconds = 0.003 # 3ms
inst_replay_frs: SpikeRateTrends = SpikeRateTrends.init_from_spikes_and_epochs(spikes_df=global_session.spikes_df, filter_epochs=global_replays, included_neuron_ids=EITHER_subset.track_exclusive_aclus.copy(), instantaneous_time_bin_size_seconds=replay_instantaneous_time_bin_size_seconds)
neuron_labels = [str(aclu) for aclu in inst_replay_frs.included_neuron_ids] # labels for plotting the epochs using `BasicBinnedImageRenderingWindow`
# 26 seconds
n_cells = np.shape(neuron_labels)
```

```python
enable_real_offset_times: bool = True # if False, indicies of time bins will be displayed along the x axis. Otherwise, real times will be used.
enable_inter_epoch_zero_padding: bool = True # if True, zeros are inserted at the start and stop of each epoch to enable correct plotting when `plot_all_epochs_sequentially` is True
padding_value = 0.0
# padding_value = np.nan

n_epochs = len(inst_replay_frs.inst_fr_df_list)

flat_inst_frs = []
flat_time_bin_times = []
_accrued_num_time_bins: int = 0
flat_epoch_separator_line_times = [] # one before/after each epoch.

nCells: int = len(inst_replay_frs.included_neuron_ids) # fixed number of cells

for epoch_idx in np.arange(n_epochs):
	a_replay_inst_frs = inst_replay_frs.inst_fr_df_list[epoch_idx].to_numpy() # (n_epoch_time_bins, n_cells)
	n_epoch_time_bins, an_epoch_n_cells = np.shape(a_replay_inst_frs)
	# display(a_replay_inst_frs.shape)
	an_epoch_peak_idxs = np.argmax(inst_replay_frs.inst_fr_df_list[epoch_idx].to_numpy(), axis=0)
	a_epoch_timebin_labels = [str(v) for v in np.arange(np.shape(a_replay_inst_frs)[0])]
	assert len(a_epoch_timebin_labels) == np.shape(a_replay_inst_frs)[0]
	assert len(neuron_labels) == np.shape(a_replay_inst_frs)[1] # n_cells
	# extract time bin times:
	an_epoch_signal = inst_replay_frs.inst_fr_signals_list[epoch_idx]
	an_epoch_time_bin_times = an_epoch_signal.times.magnitude # .magnitude just drops the units. array([43.4297, 43.4327, 43.4357, ..., 43.4837, 43.4867, 43.4897, 43.4927])
	assert len(an_epoch_time_bin_times) == n_epoch_time_bins
	an_epoch_start_t, an_epoch_end_t = an_epoch_time_bin_times[0], an_epoch_time_bin_times[-1]
	if enable_inter_epoch_zero_padding:
		# one sample ahead of the current epoch:
		flat_time_bin_times.append(np.array([(an_epoch_start_t-inst_replay_frs.instantaneous_time_bin_size_seconds)]))
		flat_inst_frs.append(np.full((1, an_epoch_n_cells), fill_value=padding_value)) # pad with one value
	
	## append the values:
	flat_time_bin_times.append(an_epoch_time_bin_times)
	flat_inst_frs.append(a_replay_inst_frs)
	if enable_inter_epoch_zero_padding:
		# one sample following it:
		flat_time_bin_times.append(np.array([(an_epoch_end_t+inst_replay_frs.instantaneous_time_bin_size_seconds)]))
		flat_inst_frs.append(np.full((1, an_epoch_n_cells), fill_value=padding_value)) # pad with one value
		
	if enable_real_offset_times:
		flat_epoch_separator_line_times.append(an_epoch_start_t)
		flat_epoch_separator_line_times.append(an_epoch_end_t)
	else:
		# indicies only:
		flat_epoch_separator_line_times.append(_accrued_num_time_bins)
		flat_epoch_separator_line_times.append((_accrued_num_time_bins+n_epoch_time_bins))
	# increment the accrued number of time bins:
	_accrued_num_time_bins += n_epoch_time_bins
	

# build flat arrays:
flat_time_bin_times = np.concatenate(flat_time_bin_times) # (n_flat_time_bins)
flat_inst_frs = np.concatenate(flat_inst_frs, axis=0) # (n_flat_time_bins, n_cells)
# validate:
n_flat_time_bins = np.shape(flat_time_bin_times)[0]
assert n_flat_time_bins == np.shape(flat_inst_frs)[0]
assert nCells == np.shape(flat_inst_frs)[1]
flat_epoch_separator_line_times = np.array(flat_epoch_separator_line_times)
assert len(flat_epoch_separator_line_times) == (n_epochs*2), f"len(flat_epoch_separator_line_times): {len(flat_epoch_separator_line_times)}, (n_epochs*2): {(n_epochs*2)}"
```
