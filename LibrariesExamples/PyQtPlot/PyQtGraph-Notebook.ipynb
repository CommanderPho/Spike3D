```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QHBoxLayout, QSlider, QLabel
from PyQt5.QtCore import Qt
import pyqtgraph as pg

class Example(QWidget):
    def __init__(self):
        super().__init__()

        self.initUI()

    def initUI(self):
        layout = QHBoxLayout(self)

        self.label = QLabel('0', self)
        layout.addWidget(self.label)

        self.slider = QSlider(Qt.Horizontal, self)
        self.slider.setMinimum(0)
        self.slider.setMaximum(100)
        self.slider.valueChanged.connect(self.updateSpinBox)

        self.spinBox = pg.SpinBox(self)
        self.spinBox.setRange(0, 100)
        self.spinBox.valueChanged.connect(self.updateSlider)

        layout.addWidget(self.slider)
        layout.addWidget(self.spinBox)

        self.setGeometry(300, 300, 300, 200)
        self.setWindowTitle('Slider and SpinBox Sync')
        self.show()

    def updateSpinBox(self, value):
        self.label.setText(str(value))
        self.spinBox.blockSignals(True)
        self.spinBox.setValue(value)
        self.spinBox.blockSignals(False)

    def updateSlider(self, value):
        self.label.setText(str(value))
        self.slider.blockSignals(True)
        self.slider.setValue(int(value))
        self.slider.blockSignals(False)


if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = Example()
    sys.exit(app.exec_())
```

```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QHBoxLayout, QSlider, QLabel
from PyQt5.QtCore import Qt
import pyqtgraph as pg

class Example(QWidget):
    def __init__(self):
        super().__init__()

        self.initUI()

    def initUI(self):
        layout = QHBoxLayout(self)

        self.label = QLabel('0', self)
        layout.addWidget(self.label)

        self.slider = QSlider(Qt.Horizontal, self)
        self.slider.setMinimum(0)
        self.slider.setMaximum(100)
        self.slider.valueChanged.connect(self.syncWidgets)
        layout.addWidget(self.slider)

        self.spinBox = pg.SpinBox(self)
        self.spinBox.setRange(0, 100)
        self.spinBox.valueChanged.connect(self.syncWidgets)
        layout.addWidget(self.spinBox)

        self.setGeometry(300, 300, 300, 100)
        self.setWindowTitle('Slider, Label and SpinBox')
        self.show()

    def syncWidgets(self, value):
        self.slider.setValue(value)
        self.spinBox.setValue(value)
        self.label.setText(str(value))

if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = Example()
    sys.exit(app.exec_())
```
